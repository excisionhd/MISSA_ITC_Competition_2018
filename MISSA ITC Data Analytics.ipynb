### Feature Binning/Mapping:
If AuditDuration <= 47, classify it as "0"  
If AuditDuration > 47 and <= 97, classify it as "1"  
If AuditDuration < 97, classify it as "2" 

These values were obtained by deviating by 1 standard deviation from the mean.  (See "Statistics about Sheet2", specifically "AuditDuration")

```python
train_sheet2 = [sheet2]
#55, 55, 80, 80
for dataset in train_sheet2:
    dataset.loc[ dataset['AuditDuration'] <= 47, 'AuditDuration'] = 0,
    dataset.loc[(dataset['AuditDuration'] > 47) & (dataset['AuditDuration'] <= 97), 'AuditDuration'] = 1,
    dataset.loc[ dataset['AuditDuration'] > 97, 'AuditDuration'] = 2
```

Define a function called bar_chart to quickly graph features based on whether or not the health inspection failed or passed.

```python
def bar_chart(feature):
    passed = sheet2[sheet2['PFFlag']=='P'][feature].value_counts()
    failed = sheet2[sheet2['PFFlag']=='F'][feature].value_counts()
    df = pd.DataFrame([passed,failed])
    df.index = ['Passed','Failed']
    ax = df.plot(kind='bar',stacked=True, figsize=(25,10), fontsize=15, rot=0)
    ax.legend(loc=1, prop={'size': 28})
```

```python
def bar_chart2(feature):
    passed = sheet2[sheet2['PFFlag']=='P'][feature].sum()
    failed = sheet2[sheet2['PFFlag']=='F'][feature].sum()
    df = pd.DataFrame([passed,failed])
    df.index = ['Passed','Failed']
    ax = df.plot(kind='bar',stacked=True, figsize=(10,10), fontsize=15, colormap='Accent',rot=0)
    ax.legend(["P18"]);
    plt.setp(ax.get_legend().get_texts(), fontsize='20') # for legend text
```

```python
bar_chart('AuditDuration')
```

```python
bar_chart2('P18')
```

### Calculate Pass_Rate and concatenate sheet2 and sheet3:
Essentially, I wanted to determine the pass rate of every StoreJoin (StoreID) in sheet2 (since some stores had several health inspections done, a percentage was necessary to determine the "pass rate" for each respective stores.  

Since Sheet3 does not have duplicate stores, concatenating sheet2 and sheet3 would not be possible; thus it would be easier to determine the pass rate of each individual store and create a new DataFrame with their respective values.

```python
s3 = sheet3.set_index(['StoreJoin'])
s3 = s3.rename(columns={'ShipmentAvgCases': 'TotalShipmentAvgCases'})
s2 = sheet2[['StoreJoin', 'PFFlag']].groupby(['StoreJoin'], as_index=True).mean()
newS3 = pd.concat([s2, s3], axis=1)
finalS3 = pd.concat([s2, s3], axis=1)
newS3 = newS3.rename(columns={'PFFlag': 'Pass_Rate'})
finalS3 = newS3.rename(columns={'PFFlag': 'Pass_Rate'})
newS3.head(20)
```
